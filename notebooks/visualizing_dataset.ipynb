```python
import pandas as pd

def unify_date_columns(df: pd.DataFrame, date_x_col: str = 'Date_x', date_y_col: str = 'Date_y') -> pd.DataFrame:
    """
    Convert the specified date columns in a DataFrame to a consistent date format (YYYY-MM-DD).

    This function:
    - Converts the given columns to datetime using `pd.to_datetime`.
    - Extracts just the date portion (YYYY-MM-DD) by using `.dt.date`.
    
    Args:
        df (pd.DataFrame): The DataFrame containing the date columns.
        date_x_col (str): The name of the 'Date_x' column in df.
        date_y_col (str): The name of the 'Date_y' column in df.
        
    Returns:
        pd.DataFrame: The modified DataFrame with both date columns standardized as date objects.
    """
    # Process Date_x column if it exists
    if date_x_col in df.columns:
        df[date_x_col] = pd.to_datetime(df[date_x_col], errors='coerce').dt.date

    # Process Date_y column if it exists
    if date_y_col in df.columns:
        df[date_y_col] = pd.to_datetime(df[date_y_col], errors='coerce').dt.date

    return df

```

```python

def plot_interactive_timeseries(df: pd.DataFrame, x: str = "Date", y: str = "Close", symbol_col: str = "symbol"):
    """
    Create an interactive timeseries line plot of stock prices using Plotly.

    Args:
        df (pd.DataFrame): DataFrame containing the time series data.
        x (str): The column name for the x-axis (usually a date).
        y (str): The column name for the y-axis (e.g., stock price).
        symbol_col (str): The column to facet or color by (e.g., symbol).
    """
    fig = px.line(df, x=x, y=y, color=symbol_col, title="Stock Prices Over Time")
    fig.update_xaxes(rangeslider_visible=True)
    fig.show()


def plot_financial_radar_chart(df: pd.DataFrame, symbol: str, metrics: list):
    """
    Plot a radar (polar) chart to visualize multiple financial metrics for a given symbol.
    
    Args:
        df (pd.DataFrame): The merged or financial DataFrame with metrics.
        symbol (str): The symbol of the company to visualize.
        metrics (list): List of metric column names to plot.
    """
    # Filter for the symbol and the latest available date
    symbol_df = df[df["symbol"] == symbol].dropna(subset=metrics)
    if symbol_df.empty:
        logging.warning(f"No data available for symbol: {symbol}")
        return

    # Take the last row (most recent)
    latest = symbol_df.iloc[-1]

    categories = metrics
    values = [latest[m] for m in metrics]

    fig = px.line_polar(r=df[metrics].iloc[-1], theta=categories, line_close=True,
                        title=f"Radar Chart for {symbol} Financial Metrics",
                        template="plotly_dark")
    fig.show()


def plot_correlation_heatmap(df: pd.DataFrame, numeric_cols: Optional[list] = None):
    """
    Plot a correlation heatmap of numeric columns in the given DataFrame.
    
    Args:
        df (pd.DataFrame): DataFrame with numeric columns.
        numeric_cols (list, optional): List of numeric columns to include. If None, all numeric cols are used.
    """
    if numeric_cols is None:
        numeric_cols = df.select_dtypes(include=np.number).columns

    corr = df[numeric_cols].corr()

    plt.figure(figsize=(12,8))
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title("Correlation Heatmap of Key Metrics")
    plt.show()


import os
import logging
import pandas as pd
from src.data.preprocess import (
    preprocess_financial_data,
    preprocess_stock_data,
    merge_financial_and_stocks,
    plot_interactive_timeseries,
    plot_financial_radar_chart,
    plot_correlation_heatmap
)

# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s %(levelname)s:%(message)s')
logger = logging.getLogger(__name__)

def main():
    financial_path = "/Users/benpfeffer/Library/Mobile Documents/com~apple~CloudDocs/Cloud_Code/Projects-Porfolio/Tech-Sector-Deep-Dive/data/processed/financial_statements_merged.csv"
    stocks_path = "/Users/benpfeffer/Library/Mobile Documents/com~apple~CloudDocs/Cloud_Code/Projects-Porfolio/Tech-Sector-Deep-Dive/data/processed/stocks.csv"
    
    # Check if files exist
    if not os.path.exists(financial_path):
        logger.error(f"Financial data file not found at {financial_path}")
        return
    if not os.path.exists(stocks_path):
        logger.error(f"Stocks data file not found at {stocks_path}")
        return
    
    # Load data
    logger.info(f"Loading financial data from: {financial_path}")
    financial_df = pd.read_csv(financial_path)
    logger.info(f"Financial data loaded successfully. Shape: {financial_df.shape}")
    logger.debug(f"Financial data head:\n{financial_df.head()}")

    logger.info(f"Loading stocks data from: {stocks_path}")
    stocks_df = pd.read_csv(stocks_path)
    logger.info(f"Stocks data loaded successfully. Shape: {stocks_df.shape}")
    logger.debug(f"Stocks data head:\n{stocks_df.head()}")
    
    # Preprocess data
    logger.info("Starting financial data preprocessing...")
    financial_clean = preprocess_financial_data(financial_df, date_col="fiscalDateEnding")
    logger.info(f"Financial data preprocessing complete. Shape: {financial_clean.shape}")
    logger.debug(f"Cleaned financial data head:\n{financial_clean.head()}")

    logger.info("Starting stocks data preprocessing...")
    stocks_clean = preprocess_stock_data(stocks_df, date_col="Date")
    logger.info(f"Stocks data preprocessing complete. Shape: {stocks_clean.shape}")
    logger.debug(f"Cleaned stocks data head:\n{stocks_clean.head()}")
    
    # Merge data
    logger.info("Merging financial and stocks data...")
    merged_df = merge_financial_and_stocks(
        financial_clean, 
        stocks_clean, 
        financial_date_col="fiscalDateEnding", 
        stock_date_col="Date"
    )
    logger.info(f"Merging complete. Shape: {merged_df.shape}")
    logger.debug(f"Merged data head:\n{merged_df.head()}")

    # Visualization examples
    logger.info("Generating interactive time series plot...")
    plot_interactive_timeseries(stocks_clean.reset_index(), x="Date", y="Close", symbol_col="symbol")
    logger.info("Time series plot generated successfully.")

    logger.info("Generating financial radar chart for AAPL...")
    plot_financial_radar_chart(merged_df, symbol="AAPL", metrics=["grossProfit", "totalRevenue", "operatingIncome", "netIncome"])
    logger.info("Radar chart generated successfully.")

    logger.info("Generating correlation heatmap...")
    plot_correlation_heatmap(merged_df, numeric_cols=["grossProfit", "totalRevenue", "operatingIncome", "netIncome", "Close"])
    logger.info("Correlation heatmap generated successfully.")

    logger.info("All steps completed successfully.")
```

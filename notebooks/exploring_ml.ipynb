```python
import pandas as pd
from src.ml.price_predictor import HybridPricePredictor, PricePredictionEnvironment

# Load your price data
data = pd.read_csv('/Users/benpfeffer/Library/Mobile Documents/com~apple~CloudDocs/Cloud_Code/Projects-Porfolio/Tech-Sector-Deep-Dive/data/processed/_AAPL.csv')
data = data.drop(columns=['Year','Date'])
```

```python


# Initialize the predictor
predictor = HybridPricePredictor()

# Preprocess the data
scaled_data, clusters, reduced_data = predictor.preprocess_data(data)

# Create the environment
env = PricePredictionEnvironment(scaled_data)

# Training loop
episodes = 100
for episode in range(episodes):
    state = env.reset()
    done = False
    
    while not done:
        action = predictor.act(state)
        next_state, reward, done, _ = env.step(action)
        predictor.remember(state, action, reward, next_state, done)
        predictor.replay()
        state = next_state
```
